#Import Libraries

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

#Data collection and processing

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
#Print 5 first rows
data_frame.head()
```

```python
#add the target column to the dataframe
```

```python
data_frame["label"] = breast_cancer_dataset.target
```

```python
data_frame.shape
```

```python
#getting information about data
data_frame.info()
```

```python
#statistical measurements on the dataset
data_frame.describe()
```

```python
#check how many positive and how many negative, 1: benign, 0: malignant
data_frame["label"].value_counts()
```

```python
data_frame.groupby("label").mean()
```

```python
X = data_frame.drop(columns="label", axis=1)
Y = data_frame["label"]
```

```python
print(X)
```

#Split training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Standardize Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

#Building the Neural Network

download.png

```python
#Import tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3) #Generate random state
from tensorflow import keras
```

```python
#Setting up layers of neural network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)), #Flatten converts data into a single-dimension array
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')#Number of neurons in the output MUST be equal to the number of classes you have. In this case, 2.
])
```

```python
#Compiling the neural network

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

#Training the neural network

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

#Visualize Accuracy and Loss

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

#Accuracy of model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

#Build a predictive system

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
